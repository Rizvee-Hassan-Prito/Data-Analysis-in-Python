```python
df["Sensor_10"]=df["Sensor_10"].fillna(df["Sensor_10"].mean())
df.head()
```

```python
df_number=df[["Sensor_1","Sensor_2","Sensor_3","Sensor_4","Sensor_5"]]
df_number=df_number.fillna(df_number.mean())
df_number.isna().sum()
```

```python
df["Sensor_Status"]=df["Sensor_Status"].fillna(df["Sensor_Status"].mode()[0])
df.isna().sum()
```

```python
df["Shift"]=df["Shift"].fillna(df["Shift"].mode())
df.isna().sum()
```

```python
df["Line"]=df["Line"].fillna(df["Line"].mode())
df.isna().sum()
```

```python
df=df[["Sensor_1","Sensor_2","Sensor_3","Sensor_4","Sensor_5", "Sensor_6","Sensor_7","Sensor_8","Sensor_9","Sensor_10"]]
df=df.fillna(df.mean())
df.isna().sum()
```
