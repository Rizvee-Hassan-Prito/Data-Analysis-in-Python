```python
def calculate_salary(horly_wage,hours_worked):
    return horly_wage * hours_worked

print("Basic salary:",calculate_salary(20,40))
```

```python
def calculate_salary_with_bonus(horly_wage,hours_worked):
  basic_salary = horly_wage * hours_worked
  if hours_worked >40:
    bonus=(hours_worked-40)*hours_worked*10
    basic_salary+=bonus
  return basic_salary
print("Basic salary:",calculate_salary_with_bonus(20,45))
```

```python
def deduct_tax(salary):
  if salary>1000:
    tax=salary*0.15
    return salary-tax
  return salary

print("Salary after tax:",deduct_tax(1200))
```

```python
def calculate_final__salary(horly_wage,hours_worked):
  salary_with_bonus=calculate_salary_with_bonus(horly_wage,hours_worked)
  final_salary=deduct_tax(salary_with_bonus)
  return final_salary

print("Final salary:",calculate_final__salary(20,50))
```

```python
def calculate_salaries_for_employees():
  num_employees=int(input("Enter number of employees:"))
  for i in range(num_employees):
    print(f"\nEmployee{i+1}")
    hourly_wage=float(input("Enter hourly wage:"))
    hours_worked=float(input("Enter hours worked:"))
    final_salary=calculate_final__salary(hourly_wage,hours_worked)
    print(f"Final Salary for employee {i+1}:${final_salary:.2f}")
calculate_salaries_for_employees()
```

```python
def classify_salary_brackets(hourly_wage,hours_worked):
  final_salary=calculate_final__salary(hourly_wage,hours_worked)
  if final_salary>1500:
    bracket="High"
  elif 1000<=final_salary<=1500:
    bracket="Medium"
  else:
    bracket="Low"
  return final_salary,bracket
hourly_wage=float(input("Enter hourly wage:"))
hours_worked=float(input("Enter hours worked:"))
final_salary,bracket=classify_salary_brackets(hourly_wage,hours_worked)
print(f"Final Salary: ${final_salary:.2f}- Bracket:{bracket}")
```
