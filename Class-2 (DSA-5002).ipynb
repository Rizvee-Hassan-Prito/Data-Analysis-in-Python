### Working with Strings

```python
first_string="Hello"
second_string= "World"
concatenated_string=first_string+" "+second_string
print(concatenated_string)
```

```python
print("First character of frst string:",first_string[0])
print("Last character of second string:",second_string[-1])
print("Middle character of concatenated string:",concatenated_string[len(concatenated_string)//2])
```

###Basic Arithmatic Operations

```python
a=10
b=5
print("Addition:",a+b)
print("Subtraction:",a-b)
print("Multiplication:",a*b)
print("Division:",a/b)
print("Floor Division:",a//b)
print("Modulus:",a%b)
print("Exponentiation:",a**b)
```

```python
x=10
y=20
print("x==y:",x==y)
print("x!=y:",x!=y)
print("x>y:",x>y)
print("x<y:",x<y)
print("x>=y:",x>=y)
print("x<=y:",x<=y)
```

###if-elif-else chain

```python
age=int(input("Enter your age:"))
if age<13:
  print("Child")
elif age<20:
  print("Teenager")
elif age<60:
  print("Adult")
else:
  print("Senior")
```

Nested if Statements

```python
number=int(input("Enter a number:"))
if number>0:
  print('positive')
  if number%2==0:
    print('even')
  else:
    print('odd')
elif number==0:
  print('Zero')
else:
  print('negative')
```

```python
i=1
while i<=10:
  print(i,end=" ")
  i+=1
  print()
```

```python
num_emp=int(input("Enter num of employee:"))
net_sal=0
for i in range(num_emp):
  salary=float(input(f"Enter salary of employee {i+1}:"))
  if salary <0:
    print("Invalid salary")
  if salary>1000 and salary<2000:
    net_sal=salary-salary*0.1
  elif salary>=2000 and salary<4000:
    net_sal=salary-salary*0.2
  elif salary>=4000:
    net_sal=salary-salary*0.3
  print("net salary ", net_sal)

```
