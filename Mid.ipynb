```python
import pandas as pd
df= pd.read_csv("/content/company_financial_dataset_with_missing.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
  if df[col].dtypes=='O':
    print(col+": ",df[col].unique())
```

```python
for col in df.columns:
  if df[col].dtypes!='O':
    df[col].fillna(df[col].median(),inplace=True) #inplace true means changing the underlying data without returning anything
  else:
    df[col].fillna(df[col].mode()[0],inplace=True)
#df.fillna(df.median(numeric_only=True),inplace=True)
#df.fillna({'':'Unknown','':'Unknown'})
df.isnull().sum()
```

##Numerical

```python
import pandas as pd
import matplotlib.pyplot as plt
numerical_columns = df.select_dtypes(include='number')
numerical_columns.plot(kind='line', figsize=(10, 6))
plt.title('Line Plots for All Numerical Columns')
plt.ylabel('Value')
plt.legend(title='Columns')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
sns.boxplot(numerical_columns)
```

```python
for column in numerical_columns.columns:
    plt.figure(figsize=(12, 8))
    sns.histplot(df[column], kde=True, bins=10)
    plt.title(f'Histogram with KDE for {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid()
    plt.show()
```
