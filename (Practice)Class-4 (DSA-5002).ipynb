#Functions

```python
def greet(name):
    print(f"Hello,{name}!")
greet("Jhon")
```

```python
def add(a,b):
  return a+b
print(add(2,3))
```

```python
def multiply_list(*args):
  new=[]
  for arg in args:
    new.append(arg*2)
  return new
multiply_list(1,2,3,4)
```
